```python
methods = [
    'GET is used to request  get  or retrive data from a specified resource.',
    'POST is used to send data to a server to create a resource or  submit data to be processed to a specified resource.',
    'The PUT method updates an instance of a resource by replacing the existing resource with the resource provided in the request body',
    'The DELETE method requests that the origin server remove the association between the target resource and its current functionality HTTP DELETE method deletes the specified resource at the origin of server. '
    'Other'
]
```

```python
methodName = ['Get','Post','Put','Delete']
```

```python
methods
```

```python
from sentence_transformers import SentenceTransformer
# initialize sentence transformer model
model = SentenceTransformer('bert-base-nli-mean-tokens')
# create sentence embeddings
sentence_embeddings = model.encode(methods)
sentence_embeddings.shape
```

```python
import faiss
```

```python
d = sentence_embeddings.shape[1]
index = faiss.IndexFlatIP(d)
```

```python
index.is_trained
```

```python
index.add(sentence_embeddings)
```

```python
index.ntotal
```

```python
k = 1
xq = model.encode(["hi "])
xq
```

```python
%%time
D, I = index.search(xq, k)  # search
print(I[0])
```
